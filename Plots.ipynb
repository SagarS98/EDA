```python
import numpy as np 
x = np.linspace(0,100,11)
y = x/2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('Twice of y Value')
```

```python
fig , ax = plt.subplots()

ax.plot(x,y,'r')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_title('Set Title')
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax2 = fig.add_axes([0.5,.15,0.4,0.3])

ax1.plot(x,y,'r')
ax2.plot(y,x,'b')
```

```python
a = np.linspace(5,25,5)
b = a/5

fig , axes = plt.subplots(nrows=2 ,ncols =1 ,figsize=(5,3))
axes[0].plot(x,y)
axes[1].plot(a,b)

plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(a,b,'blue',lw = 1.755, alpha =0.5, linestyle = '--',marker = 'o',ms = 25 ,mfc= 'cyan')
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z= x**2
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('Title')
```

```python
fig = plt.figure()

ax1 =fig.add_axes([0,0,1,1])

ax2 =fig.add_axes([0.2,0.5,.2,.2])

ax1.plot(x,y)

ax2.plot(x,y,'r')
```

```python
fig ,ax = plt.subplots(nrows=1,ncols=2,figsize =(12,2))

ax[0].plot(x,y,color='blue',ls ='--',lw =3,)

ax[1].plot(x,z,color='red',lw = 3,ls='--')
```

```python
import seaborn as sns

%matplotlib inline
```

```python
import scipy.stats as stat
```

```python
tips =sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.set_style('darkgrid')
sns.distplot(tips['total_bill'], kde=False)
    
    
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg',stat_func=stat.pearsonr )
```

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'] , kde =False)
```

```python
import numpy as np
```

```python
sns.catplot(x= 'sex' ,y='total_bill' ,data = tips , kind ='bar' , estimator=np.std)
```

```python
sns.boxplot(x ='day' , y ='total_bill' ,data= tips)
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
import scipy.stats as stats
sns.jointplot(x='fare',y='age',data=titanic).annotate(stats.pearsonr)
```

```python
sns.distplot(titanic['fare'],kde=False,color='red')
```

```python
sns.boxplot(x='class' ,y='age',data=titanic,palette='rainbow')
```

```python
sns.swarmplot(x='class',y='age',data=titanic,palette='rainbow')
```

```python
sns.heatmap(titanic.corr() ,cmap='coolwarm')
plt.title('titanic.corr')
```

```python
g=sns.FacetGrid(data=titanic,col='sex')
g.map(plt.hist,'age')
g.map(sns.distplot,'age' ,kde=False)
```
